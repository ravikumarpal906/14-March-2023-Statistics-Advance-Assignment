**Q1. Write a Python function that takes in two arrays of data and calculates the F-value for a variance ratio test. The function should return the F-value and the corresponding p-value for the test.**

```python
import numpy as np
from scipy.stats import f

def f_test(arr1,arr2):
  var1 = np.var(arr1)
  var2 = np.var(arr2)

  df1 = len(arr1) - 1
  df2 = len(arr2) - 1

  f_value = var1/var2

  p_value = 1 - f.cdf(f_value, df1, df2)

  return f_value, p_value

data1 = np.array([1, 2, 3, 4, 5])
data2 = np.array([2, 4, 6, 8, 10])

F , p_value = f_test(data1, data2)
print(F)
print(p_value)
```

**Q2. Given a significance level of 0.05 and the degrees of freedom for the numerator and denominator of an F-distribution, write a Python function that returns the critical F-value for a two-tailed test.**

```python
from scipy.stats import f

def critical_f_value(alpha, df_num, df_den):
  f_critical_1 = f.ppf(1 - alpha/2, df_num, df_den) # Upper critical value
  f_critical_2 = f.ppf(alpha/2, df_num, df_den) #Lower critical value

  return f_critical_1, f_critical_2

alpha = 0.05
df_num = 10
df_den = 20

f_critical_1, f_critical_2 = critical_f_value(alpha, df_num, df_den)
print("Upper critical value",f_critical_1)
print("Lower critical value ",f_critical_2)
```

**Q3. Write a Python program that generates random samples from two normal distributions with known variances and uses an F-test to determine if the variances are equal. The program should output the F-value, degrees of freedom, and p-value for the test.**

```python
import numpy as np
from scipy.stats import f

def f_test(arr1 ,arr2):
  var1 = np.var(arr1)
  var2 = np.var(arr2)

  df1 = len(arr1) - 1
  df2 = len(arr2) - 1

  F = var1/var2

  p_value =f.sf(F, df1, df2)

  return F, df1, df2, p_value

mean1 , mean2 = 0,0
var1 , var2 = 1,1.5
sample_size = 100

np.random.seed(42)
data1 = np.random.normal(mean1, var1, sample_size)
data2 = np.random.normal(mean2, var2, sample_size)

F , df1 , df2 , p_value = f_test(data1, data2)
# Print the results
print("F-value: ", F)
print("Degrees of freedom for sample1: ", df1)
print("Degrees of freedom for sample2: ", df2)
print("p-value: ", p_value)
```

**Q6. Write a Python function that takes in the degrees of freedom for the numerator and denominator of an F-distribution and calculates the mean and variance of the distribution. The function should return the mean and variance as a tuple.**

```python
# prompt: Q6. Write a Python function that takes in the degrees of freedom for the numerator and denominator of an
# F-distribution and calculates the mean and variance of the distribution. The function should return the
# mean and variance as a tuple.

def f_distribution_properties(df_num, df_den):
  """
  Calculates the mean and variance of an F-distribution.

  Args:
      df_num: Degrees of freedom for the numerator.
      df_den: Degrees of freedom for the denominator.

  Returns:
      A tuple containing the mean and variance of the F-distribution.
  """

  if df_den <= 2:
      raise ValueError("Denominator degrees of freedom must be greater than 2.")

  mean = df_den / (df_den - 2)
  variance = (2 * df_den ** 2 * (df_num + df_den - 2)) / (df_num * (df_den - 2) ** 2 * (df_den - 4))

  return mean, variance

df_num = 10
df_den = 20

mean, variance = f_distribution_properties(df_num, df_den)

print("Mean:", mean)
print("Variance:", variance)
```

**Q7. A random sample of 10 measurements is taken from a normal population with unknown variance. The sample variance is found to be 25. **Q8. The following data represent the waiting times in minutes at two different restaurants on a Saturday night: Restaurant A: 24, 25, 28, 23, 22, 20, 27; Restaurant B: 31, 33, 35, 30, 32, 36.
